```python
data_0_1 = data[data['gesture'] == '0_1']
data_1_0 = data[data['gesture'] == '1_0']
data_0_1['sample_num'] = data_0_1['sample_num']+len(data_1_0['sample_num'].unique())+1
data = pd.concat([data_1_0, data_0_1], axis = 0)
data_1_0
```

```python
N=10
nums = [[0 for i in range(N)]]
for i in range(1,int(100/N)):
    nums.append([i for j in range(N)])
    nums
nums = np.array(nums).reshape((-1,))
rebuild_data = []
for sample in data['sample_num'].unique():
    names_list = []
    sample_by_data = data[data['sample_num'] == sample]
    sample_by_data['sub_period'] = nums
    accs = pd.Series()
    for part in sample_by_data['sub_period'].unique():
        part_data = sample_by_data[sample_by_data['sub_period']==part]
        for i in range(14):
            accs['mean_'+str(i)+"_"+str(part)] = part_data[str(i)].mean(axis = 0)
            accs['min_'+str(i)+"_"+str(part)] = part_data[str(i)].min(axis = 0)
            accs['max_'+str(i)+"_"+str(part)] = part_data[str(i)].max(axis = 0)
            names_list.append('mean_'+str(i)+"_"+str(part))
            names_list.append('min_'+str(i)+"_"+str(part))
            names_list.append('max_'+str(i)+"_"+str(part))
    names_list.append('gesture')
    accs['gesture'] = sample_by_data['gesture'].unique()[0]
    rebuild_data.append(accs.values)
rebuild_data = pd.DataFrame(rebuild_data)
rebuild_data.columns = names_list
rebuild_data
rebuild_data['gesture'] = le.transform(rebuild_data['gesture'])
```

```python
rebuild_data.to_csv('full_data_transformed.csv')
```

```python
X = rebuild_data.drop(['gesture'], axis = 1)
y = rebuild_data['gesture']
X
```

```python
from sklearn.model_selection import train_test_split

X_train,  X_control, y_train, y_control = train_test_split(X, y, shuffle = True, test_size=0.1)
```
